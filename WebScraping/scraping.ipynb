```python
# Importing Dependencies.
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# List to store all data.
all_data = []

# Using a for loop since the page has multiple pages.
for page_num in range(1, 7):
    url = (
        f"https://www.scrapethissite.com/pages/forms/?page_num={page_num}&per_page=100"
    )

    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    # Finding the table
    table = soup.find("table")

    # Finding the table rows
    rows = table.find_all("tr")

    # Extracting the actual table data
    for row in rows[1:]:
        cols = row.find_all("td")

        individual_row_data = [col.text.strip() for col in cols]

        all_data.append(individual_row_data)
```

```python
headers = table.find_all("th")
cleaned_headers = [header.text.strip() for header in headers]
df = pd.DataFrame(all_data, columns=cleaned_headers)
```
